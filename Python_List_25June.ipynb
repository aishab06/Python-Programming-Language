Question 3: Write a Python program to remove the first occurrence of a specified element from a list.
- sample input :
var = [1,2,3,4,2,5]
- sample output :
result : [1,3,4,2,5]

```python
var = [1,2,3,4,2,5]
newVar = {}
for i in var:
  if i in newVar:
    newVar[i] = newVar[i] + 1
  else:
    newVar[i] = 1

newLst = []
removed_already = {}
for item in var:
  if newVar[item] > 1 and item not in removed_already:
    removed_already[item] = True
    continue
  newLst.append(item)

print (f'result: {newLst}')
```

Question 4: Write a Python program to find the average of all the elements in a list.
- sample input :
var = [1,2,3,4,2,5]
- sample output :
result : 2.833

```python
var = [1,2,3,4,2,5]
count = 0
sumNum = 0
for i in var:
  count = count + 1
  sumNum = sumNum + i
avrg = sumNum / count
print (f'result: {avrg:.3f}')
```

Question 7: Write a Python program to find which number is palindrome in below given list .
- sample input :
var = [134,121,155]
- sample output :
result : 134 is not a palindome , 121 is a palindrome

```python
var = [134,121,155]
result_lst = []
for item in var:
  reversed_num = 0
  n = item
  while n > 0:
    digit = n %10
    reversed_num = (reversed_num * 10) + digit
    n = n // 10
  if item == reversed_num:
    status = 'is a palindrome'
  else:
    status = 'is not a palindrome'
  message = f'{item} {status}'
  result_lst.append(message)
print ('result:', ', '.join(result_lst))

#  if item == reversed_num:
#    result_lst.append(f'{item} is palindrome')
#    #print (f'result: {item} is palindrome')
#  else:
#    result_lst.append(f'{item} is not a palindrome')
#    #print (f'result: {item} is not a palindrome')
#print (result_lst)
```

Question 8: Write a Python program to find the common elements between two lists.
- sample input :
var1 = [1,2,3,4,10]
var2 = [10,11,12,14]
- sample output :
result : 10

```python
var1 = [1,2,3,4,10]
var2 = [10,11,12,14]
new_lst = []
for i in var1:
  if i in var2:
    new_lst.append(i)
result = ','.join(str(n) for n in new_lst)
print(f'result: {result}')
```

Question 9: Write a Python program to remove the duplicates from a list
- sample input :
var = [10,11,12,10,12,24]
- sample output :
result : [10,11,12,24]

```python
var = [10,11,12,10,12,24]
dupl_lst = set(var)
print (f'result: {list(dupl_lst)}')

newLst = []
for i in var:
  if i not in newLst:
    newLst.append(i)
print (f'result: {newLst}')
```

Question 10: Write a Python program to sort a list of strings in Ascii Order [Ascending order]
- sample input :
var = ['a' , 'c' , 'b']
- sample output :
result : ['a' , 'b' , 'c']

```python
var = ['a' , 'c' , 'b', 'e']
asci_lst = {}
for item in var:
  asci_lst[item] = ord(item)
print (asci_lst)
sorted_keys = sorted(asci_lst, key=asci_lst.get)
print (f'result: {sorted_keys}')
```
