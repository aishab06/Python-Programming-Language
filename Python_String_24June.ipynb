Question 4: Write a Python program to find the most common character in a string.
- sample input :
var = 'Hello world '
- sample output :
result : l, because l is repeating 3 times

```python
var = 'Hello woereled'
char_count = {}
for char in var:
  if char in char_count:
    char_count[char] = char_count[char] + 1
  else:
    char_count[char] = 1

common = max(char_count, key=char_count.get)
print (f'result: {common}, because {common} is repeating {char_count[common]} times')
```

Question 5: Write a Python program to check if two strings are anagrams.
- sample input :
var1 = 'listen'
var2 = 'silent'
- sample output :
result : both are anagram because characters in var2 should present in var1

```python
var1 = 'listen'
var2 = 'silent'
print (f'result: both are anagram') if sorted(var1) == sorted(var2) else print ('result: both are not anagram')

count1 = {}
count2 = {}
for char in var1:
  if char in count1:
    count1[char] = count1[char] + 1
  else:
    count1[char] = 1

for char in var2:
  if char in count2:
    count2[char] = count2[char] + 1
  else:
    count2[char] = 1

print (count1, count2)
if count1 == count2:
  print ('result: both are anagram')
else:
  print ('result: both are not anagram')

```

Question 7: Write a Python program to find the longest word in a string.
- sample input :
var = 'python is Easy Language'
- sample output :
result : Language

```python
var = 'python is Easy Language'
spWord = var.split(' ')
wrdCount = {}
for wordLst in spWord:
  wrdCount[wordLst] = len(wordLst)
maxCnt = max(wrdCount, key=wrdCount.get)
print (f'result: {maxCnt}')
```

Question 8: Write a Python program to capitalize the first letter of each word in a string [dont use built-in-function called title].
- sample input :
var = 'python is easy'
- sample output :
result : Python Is Easy

```python
var = 'python is easy'
spWord = var.split(' ')
newLst = []
for wordLst in spWord:
  newLst.append(wordLst.capitalize())

print (f'result: {' '.join(newLst)}')
```

Question 10: write a python programme to find the sum of all the even characters based on Ascii values
- sample input :
var = 'sharuk khan'
- sample output :
result : 464

```python
var = 'sharuk khan'
ascLst=[]
res = 0
for i in var:
  ascLst.append(ord(str(i)))

for item in ascLst:
  if item % 2 == 0:
    res = res + item

print (f'result: {res}')

```
